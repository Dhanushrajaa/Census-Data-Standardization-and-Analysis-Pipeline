```python
import pandas as pd
```

```python
#Task 1
df=pd.read_excel("census_2011.xlsx")
renaming_col={
    'District code':'District_code',
    'State name': 'State_UT',
    'District name': 'District',
    'Male_Literate': 'Literate_Male',
    'Female_Literate': 'Literate_Female',
    'Rural_Households': 'Households_Rural',
    'Urban_Households': 'Households_Urban',
    'Age_Group_0_29': 'Young_and_Adult',
    'Age_Group_30_49': 'Middle_Aged',
    'Age_Group_50': 'Senior_Citizen',
    'Age not stated': 'Age_Not_Stated'
}
df.rename(columns=renaming_col,inplace=True)

```

```python
df.columns
```

```python
#Task 2
# Function to standardize State/UT names
def standardize_state_name(state_name):
    words = state_name.split()
    standardized_words = [word.capitalize() if word.lower() != 'and' else word.lower() for word in words]
    return ' '.join(standardized_words)

# Apply the function to the 'State/UT' column
df['State_UT'] = df['State_UT'].apply(standardize_state_name)
```

```python
df
```

```python
#Task 3
# List of districts that moved to Telangana
telangana_districts = [
    "Adilabad", "Nizamabad", "Karimnagar", "Medak", "Hyderabad",
    "Rangareddy", "Mahbubnagar", "Nalgonda", "Warangal", "Khammam"
]

# List of districts that moved to Ladakh
ladakh_districts = ["Leh(Ladakh)", "Kargil"]

# Update the State/UT column based on the districts
df.loc[df['District'].isin(telangana_districts), 'State_UT'] = 'Telangana'
df.loc[df['District'].isin(ladakh_districts), 'State_UT'] = 'Ladakh'
```

```python
   
# selecting rows based on condition 
rslt_df = df[df['State_UT'].isin(["Ladakh","Telangana"])] 
   
print('\nResult :\n',
      rslt_df)
```

```python
#Task 4

# Calculate the percentage of missing data for each column
missing_data_percentage_before = df.isnull().mean() * 100

# Print the percentage of missing data before filling
#print("Percentage of missing data before filling:")
#print(missing_data_percentage_before)

# Fill missing Population using Male + Female
df['Population'] = df['Population'].fillna(df['Male'] + df['Female'])
df['Male'] = df['Male'].fillna(df['Population'] - df['Female'])
df['Female'] = df['Female'].fillna(df['Population'] - df['Male'])

# Fill missing Literate using Literate_Male + Literate_Female
df['Literate'] = df['Literate'].fillna(df['Literate_Male'] + df['Literate_Female'])
df['Literate_Male'] = df['Literate_Male'].fillna(df['Literate'] - df['Literate_Female'])
df['Literate_Female'] = df['Literate_Female'].fillna(df['Literate'] - df['Literate_Male'])

# Fill missing Population using age groups
df['Population'] = df['Population'].fillna(df['Young_and_Adult'] + df['Middle_Aged'] + df['Senior_Citizen'] + df['Age_Not_Stated'])
df['Young_and_Adult'] = df['Young_and_Adult'].fillna(df['Population'] - df['Middle_Aged'] - df['Senior_Citizen'] - df['Age_Not_Stated'])
df['Middle_Aged'] = df['Middle_Aged'].fillna(df['Population'] - df['Young_and_Adult'] - df['Senior_Citizen'] - df['Age_Not_Stated'])
df['Senior_Citizen'] = df['Senior_Citizen'].fillna(df['Population'] - df['Young_and_Adult'] - df['Middle_Aged'] - df['Age_Not_Stated'])
df['Age_Not_Stated'] = df['Age_Not_Stated'].fillna(df['Population'] - df['Young_and_Adult'] - df['Middle_Aged'] - df['Senior_Citizen'])

# Fill missing Households using Households_Rural + Households_Urban
df['Households'] = df['Households'].fillna(df['Households_Rural'] + df['Households_Urban'])
df['Households_Rural'] = df['Households_Rural'].fillna(df['Households'] - df['Households_Urban'])
df['Households_Urban'] = df['Households_Urban'].fillna(df['Households'] - df['Households_Rural'])


# Recalculate the percentage of missing data after filling
missing_data_percentage_after = df.isnull().mean() * 100

# Print the percentage of missing data after filling
#print("\nPercentage of missing data after filling:")
#rint(missing_data_percentage_after)

missing_data_comparison = pd.DataFrame({"Before": missing_data_percentage_before,"After": missing_data_percentage_after})
print(missing_data_comparison)

# Save the updated DataFrame to a new Excel file
df.to_excel('census_data_filled.xlsx', index=False)
```

```python
import pymongo
import psycopg2

# Connect to MongoDB
mongo_client = pymongo.MongoClient("mongodb+srv://dhanushrajaa4488:1234@cluster0.gbhznoq.mongodb.net/?retryWrites=true&w=majority&appName=Cluster0")
mongo_db = mongo_client["test"]
mongo_collection = mongo_db["census"]

# Connect to PostgreSQL
postgres_conn = psycopg2.connect(
    dbname="census_db",
    user="postgres",
    password="postgresql",
    host="localhost",
    port="5432"
)
postgres_cur = postgres_conn.cursor()

# Define the PostgreSQL table schema dynamically
table_name = "census"
table_schema = f"CREATE TABLE IF NOT EXISTS {table_name} (id SERIAL PRIMARY KEY"

# Get a sample document to extract column names and types
sample_document = mongo_collection.find_one()

for key, value in sample_document.items():
    # Sanitize column name and convert to lowercase
    column_name = ''.join(c if c.isalnum() else '_' for c in key).lower()

    # Adjust data types as needed based on your MongoDB values
    postgres_type = "TEXT" if isinstance(value, str) else "INT" if isinstance(value, int) else "FLOAT" if isinstance(value, float) else "TEXT"

    # Add column to the schema
    table_schema += f", {column_name} {postgres_type}"

# Finish the schema creation
table_schema += ");"
postgres_cur.execute(table_schema)
postgres_conn.commit()

# Fetch data from MongoDB and insert into PostgreSQL
for document in mongo_collection.find():
    # Transform MongoDB document to match PostgreSQL schema
    postgres_data = {key.lower(): str(value) if isinstance(value, (list, dict)) else value for key, value in document.items() if key != '_id'}

    # Sanitize column names
    sanitized_columns = [''.join(c if c.isalnum() else '_' for c in key).lower() for key in postgres_data.keys()]

    # Create a list of column names and values for PostgreSQL insertion
    columns = ', '.join(sanitized_columns)
    placeholders = ', '.join(['%s'] * len(postgres_data))
    query = f"INSERT INTO {table_name} ({columns}) VALUES ({placeholders})"

    # Insert data into PostgreSQL
    postgres_cur.execute(query, list(postgres_data.values()))

# Commit changes and close connections
postgres_conn.commit()
postgres_cur.close()
postgres_conn.close()
mongo_client.close()
```
